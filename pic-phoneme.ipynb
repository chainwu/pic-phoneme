```python
import pandas as pd
import glob
import tgt
import re
```

```python
csvf=glob.glob('data/NSB_I_PicNaming/*naming/**/題序*/*.csv', recursive=True)
cxz=[ x for x in csvf if "Biling" not in x ]
#print(cxz)
```

```python
for cs in cxz:
    print(cs)
    pdf=pd.read_csv(cs)
    #data_top = pdf.columns.values
    img=pdf["ImageFile"].dropna().values.tolist()
    wav=pdf["mic.clip"].dropna().values.tolist()
    for i, w in zip(img, wav):
        pattern = r'[./\\]'
        iresult = re.split(pattern, i)
        wresult = re.split(pattern, w)
        
        print(iresult, wresult)
```

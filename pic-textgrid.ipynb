```python
class participants():
    def mkpartdir(self):
        #self.print()
        for i in self.indolst + self.thailst + self.vietlst:
            #print(i[0], i[1])
            nat=i[0]
            pid=i[1]
            if not os.path.exists('result/'+nat+'/'+pid): 
                os.makedirs('result/'+nat+'/'+pid) 

    def iterate_thai(self):
        self.thailst=[]
        for idx in self.thai.index:
            self.thailst.append(['thai', str(self.thai['Participant No.'][idx]).zfill(3), self.thai['Gender'][idx], self.thai['Age'][idx]])
    
    def iterate_indo(self):
        self.indolst=[]
        for idx in self.indo.index:
            self.indolst.append(['indo', str(self.indo['Participant No.'][idx]).zfill(3), self.indo['Gender'][idx], self.indo['Age'][idx]])

    def iterate_viet(self):
        self.vietlst=[]
        for idx in self.viet.index:
            self.vietlst.append(['viet', str(self.viet['Participant No.'][idx]).zfill(3), self.viet['Gender'][idx], self.viet['Age'][idx]])

    
    def __init__(self):
        self.indo = pd.read_excel('data/Participant_Info.xlsx', sheet_name='Indo')
        self.thai = pd.read_excel('data/Participant_Info.xlsx', sheet_name='Thai')
        self.viet = pd.read_excel('data/Participant_Info.xlsx', sheet_name='Viet')

        convert_dict = {'Participant No.': int, 'Age': int}
 
        self.indo=self.indo[1:-3]
        self.thai=self.thai[1:-3]
        self.viet=self.viet[1:-3]
        
        self.indo = self.indo.astype(convert_dict)
        self.thai = self.thai.astype(convert_dict)
        self.viet = self.viet.astype(convert_dict)
        
        self.iterate_thai()
        self.iterate_indo()
        self.iterate_viet()
        self.mkpartdir()
        
    def print(self):
        print(self.indolst)
        print(self.thailst)
        print(self.vietlst)

p=participants()
p.print()
```

```python
class indq():
    def __init__(self, file):
        self.df = pd.read_csv(file)
        
    def get_images(self):
        ilst = self.df['ImageFile'].tolist()
        cilst=[x for x in ilst if str(x) != 'nan']
        cxlist=[x.replace('.jpg','').replace('.jpeg','').replace('picture/','').replace('.png','').replace(' ','').split('to') for x in cilst ]
        #print(cxlist)
        finlist = [x[0] if len(x) == 1 else x[1]  for x in cxlist]
        return (finlist)

    def get_wavs(self):
        ilst = self.df['mic.clip'].tolist()
        #cxlist=[x.replace(' ') for x in ilst ]
        clst=[x for x in ilst if str(x) != 'nan']
        slst=[x.split("\\") for x in clst]
        wlist=[x[-1] for x in slst]

        #print(cxlist)
        #finlist = [x[0] if len(x) == 1 else x[1]  for x in cxlist]
  
        return (wlist)
        
    def print(self):
        il = self.get_images()
        wl = self.get_wavs()

        for i, w in zip(il, wl):
            print(i, w)

qs= questions()
for iqcsv in qs.q:
    print(iqcsv)
    iq = indq(iqcsv)
    iq.print()
    print("_______________")
```

```python
def change_participant(txtgf, pat):
    ttg=tgt.io.read_textgrid(txtgf)
    ttr=ttg.get_tier_by_name("participant")
    ann=ttr.get_annotations_with_matching_text("NAX")
    ann[0].text=pat
    tgt.io.write_to_file(ttg, txtgf, format="long")
    
```

```python
def main():
    pt=participants()
    qn=questions()
    for p in pt.thailst + pt.indolst + pt.vietlst:
        nat = p[0]
        pid = p[1]
        sex = p[2]
        age = p[3]
        destdir="./result/"+nat+"/"+pid+"/"
        srcdir="data/NSB_I_PicNaming/"+nat.capitalize()+"_Pic-naming/"+nat.capitalize()+"_PicNaming_audio/"+pid+"/"
        print(srcdir, destdir)
        print(nat, pid, destdir)
        qcsv=qn.get_csv(pid)
        if qcsv != None:
            idq=indq(qcsv)
            idimg = idq.get_images()
            idwav = idq.get_wavs()
            for img, wav in zip(idimg, idwav):
                srctxtgrid="./data/NSB_I_PicNaming/Standard_Pic-naming/"+img+".13"
                dsttxtgrid=destdir+img+".textgrid"
                print(srctxtgrid, dsttxtgrid)
                shutil.copyfile(srctxtgrid, dsttxtgrid)
                change_participant(dsttxtgrid, nat.capitalize()+" "+sex+" "+pid)
                #print(img, wav, txtgrid)
                srcwfile=srcdir+wav
                dstwfile=destdir+img+".wav"
                print(srcwfile, dstwfile)
                shutil.copyfile(srcwfile, dstwfile)
                #shutil.copyfile(wav, 'result/')
        else:
            print("cannot find qcsv")
        
main()
```
